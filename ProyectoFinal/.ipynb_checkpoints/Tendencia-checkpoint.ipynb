# Preprocesamiento

```python
stopwords2 = ('http','debate','caribe','youtube','be','youtu','https')

def getTheWordBag(archivo):
    corpus = []
    for element in archivo:
        w = re.sub('(http(s?)://)(w{3})(\.\w+)(\.(\w{3})(\.\w{2})?)(/.*)?(\s)|(<.*>)',' ',element)
        w = re.sub('(http(s?)://)?(\w+\.)?(\w+\.\w{3}(/.+)?)',' ',w)
        w = re.sub('#\w+',' ',w)
        x = re.sub('[^a-zA-Zóáéíúñ]',' ',w).lower().split()
        corpus.append([word for word in x if (not word in set(stopwords.words('spanish'))) and (not word in stopwords2)])
    return corpus
  
```

## Se crean los tokens y los vectores de Embeddings

```python
corpus2 = []
for word in corpus:
  corpus2.append(' '.join(word))
  
Tokens = []
for x in corpus2:
  w  = nlp(x)
  t = [[t.text,t.vector] for t in w]
  if t != []:
    Tokens.append(t)
    
```

```python
allEmbeddings = []
for vector in Tokens:
  for vec in vector:
    allEmbeddings.append(vec[1])
allEmbeddings = np.array(allEmbeddings)
```

```python
print(allEmbeddings.shape)
```

## Se hace el Mapa Autoorganizado

```python
from minisom import MiniSom
map_dim = 100
som = MiniSom(map_dim,map_dim, 384) # initialization of map_dimXmap_dim SOM
som.random_weights_init(allEmbeddings)
print("Training...")
som.train_random(allEmbeddings, 1000) # trains the SOM with 1000 iterations
print ("...ready!")
```

```python
def findName(index):
  index = np.array(index)
  for x in Tokens:
    for y in x:
      if np.all(y[1] ==index):
        return y[0]
    
```

```python
import matplotlib.pyplot as plt
import random
%matplotlib inline

plt.figure(figsize=(20, 20))
for x in allEmbeddings:
  w = som.winner(x)
  text = findName(x)
  plt.text(w[0],w[1]+np.random.rand()*.9, s = text,color=(random.random(), random.random(), random.random()))
plt.grid()
plt.xlim([0, 50])
plt.ylim([0, 20])
#plt.plot()
```

```python
from bokeh.io import show,output_notebook
from bokeh.plotting import figure
```

```python
X = []
Y = []
words = []
X2 = []
t0 = time()
for x in allEmbeddings:
  win = som.winner(x)
  X2.append(win)
  X.append(win[0])
  Y.append(win[1]+np.random.rand()*.9)
  words.append(findName(x))
t1 = time()
print("el tiempo estimado es: {} minutos".format((t1-t0)/60))
```

```python
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource, HoverTool,PanTool,WheelZoomTool

source = ColumnDataSource(data=dict(x = X,y = Y,pal = words))

output_notebook()
p = figure(x_range=(0, 100),plot_height=600, plot_width=700, tools=[PanTool(),WheelZoomTool()])
p.scatter(x='x', y='y', size=10,fill_alpha=0.7,hover_alpha=0.8,hover_color="green",source=source)
p.xaxis.axis_label = 'X'
p.yaxis.axis_label = 'Y'


hover = HoverTool(tooltips=[("Palabra", "@pal")],mode="mouse", point_policy="follow_mouse")
p.add_tools(hover)

show(p)
```

# Probando Algoritmo de asociación de reglas

```python
corpus2 = []
for word in corpus:
  corpus2.append(' '.join(word))
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus2).toarray()
```

```python
try:
  import mlxtend
except ImportError:
  !pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

dataX = df = pd.DataFrame(X)
frequent = apriori(dataX, min_support=0.07, use_colnames=True)
```

```python
w = cv.get_feature_names()
print(w[2649])
print("----------------------")
#print("{}".format(frequent))
print("support     itemsets")
for x in range(len(frequent.itemsets)):
  print("{}            {}".format(frequent.support[x],w[frequent.itemsets[x][0]]))
```

```python
rules = association_rules(frequent, metric="lift", min_threshold=1)
rules.head()
```

## Usando KMEANS sobre el resultado del mapa autoorganizado


```python
X2 = np.array(X2, dtype='float64')
print(X2.shape)
print(X2)
```

```python
X_train = X2[:10000]
X_test = X2[10000:]
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X2)
```

```python
plt.scatter(X2[y_kmeans == 0, 0], X2[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X2[y_kmeans == 1, 0], X2[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X2[y_kmeans == 2, 0], X2[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(X2[y_kmeans == 3, 0], X2[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X2[y_kmeans == 4, 0], X2[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.legend()
plt.show()
```
